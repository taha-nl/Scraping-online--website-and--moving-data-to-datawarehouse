```python
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import undetected_chromedriver as uc
import re
import csv
from itertools import zip_longest
```

```python
if __name__=='__main__':
    

    browser=uc.Chrome()

    browser.maximize_window()
    time.sleep(5)
   
    for i in range(1,100):

        try:
            URL=f"https://www.avito.ma/fr/maroc/immobilier-%C3%A0_vendre?o={i}"
            browser.get(URL)
            time.sleep(1)

            browser.find_element(By.XPATH,('//button[@class="sc-1nre5ec-4 hkCerg"][1]')).click()
            time.sleep(2)

            prices=browser.find_elements(By.XPATH,('//span[@class="sc-1x0vz2r-0 izsKzL oan6tk-15 cdJtEx"]'))
            names=browser.find_elements(By.XPATH,('//h3[@class="sc-1x0vz2r-0 ezkocc oan6tk-16 hlwSBL"]'))
            types=browser.find_elements(By.XPATH,('//p[@class="sc-1x0vz2r-0 iEJWiq oan6tk-18 bhnuSP"]'))
            location=browser.find_elements(By.XPATH,('//span[@class="sc-1x0vz2r-0 hCOOjL"]'))[1::2]
            links=browser.find_elements(By.XPATH,('//a[@class="oan6tk-1 fFOxTQ"]'))

            prices_text=[price.text for price in prices]
            names_text=[name.text for name in names]
            types_text=[type.text for type in types]
            location_text=[location.text for location in location]
            link_text=[link.get_attribute('href') for link in links]
            features=[]
            descriptions=[]
            for link in link_text:
                browser.get(link)
                time.sleep(1)
                keys=[key.text for key in browser.find_elements(By.XPATH,('//li[@class="sc-qmn92k-1 ldnQxr"]/span[1]'))]
                values=[value.text for value in browser.find_elements(By.XPATH,('//li[@class="sc-qmn92k-1 ldnQxr"]/span[2]'))]
                features.append(dict(zip(keys,values)))
                try:
                    descriptions.append(browser.find_element(By.XPATH,('//p[@color="midnight_lighter"]')).text)
                except:
                    descriptions.append(None)    
            kind=[x['Type'] if 'Type'in x.keys() else None for x in features]
            Secteur=[x['Secteur'] if 'Secteur' in x.keys() else None for x in features]
            Salons=[x['Salons'] if 'Salons' in x.keys() else None for x in features]
            age=[x['Âge du bien'] if 'Âge du bien' in x.keys() else None for x in features]
            etage=[x['Étage'] if 'Étage' in x.keys() else None for x in features]
            frais_syndic=[x['Frais de syndic / mois'] if 'Frais de syndic / mois' in x.keys() else None for x in features]
            surface=[x['urface habitable'] if 'urface habitable' in x.keys() else None for x in features]

            info= [names_text, prices_text, types_text, location_text,descriptions, kind, age, Secteur, Salons,etage,frais_syndic,surface]
            B_data = zip_longest(*info)
            with open('immobilier_avito-final.csv', 'a',encoding="utf-8") as myFile:
                wr = csv.writer(myFile)
                if i==1:
                    wr.writerow(["name", "price", "type", "location","descriptions", "kind",
                        "age", "Secteur", "Salons","etage","frais_syndic","surface"])
                wr.writerows(B_data)
        except:
            continue        

        
        
    
        
        

```
